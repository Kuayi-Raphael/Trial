## Explaining the method of handling null values:

The mean was used to replace the null values and missing data.This was to cater for better analysis from data.

mean_fa

# Heatmap of correlation


```python
dataCorrelation=data.corr()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(dataCorrelation,annot=True,cmap="Blues",linewidths=10,ax=ax, cbar_kws={'label': 'Correlation'},xticklabels=45, annot_kws={"fontsize": 8})
plt.title("Data Correlation Heatmap",fontdict={"fontsize": 16})
plt.show()
```

# Subplots of pH and citric acid by quality

```python

fig, ax = plt.subplots(2, 1, figsize=(6, 8))
ax[0].scatter(data['pH'], data['quality'], alpha=0.2)
ax[0].set_xlabel('pH')
ax[0].set_ylabel('Quality')
ax[0].set_title('pH vs Quality')
ax[1].scatter(data['citric acid'], data['quality'], alpha=0.2)
ax[1].set_xlabel('Citric Acid')
ax[1].set_ylabel('Quality')
ax[1].set_title('Citric Acid vs Quality')
plt.tight_layout()
plt.show()
```

# Bar graph of alcohol content by quality

```python

mean_alcohol = data.groupby('quality')['alcohol'].mean()
plt.bar(mean_alcohol.index, mean_alcohol)
plt.title('Average Alcohol Content by Quality')
plt.xlabel('Quality')
plt.ylabel('Alcohol (%)')
plt.show()
```

# Pie chart of distribution of wine quality rating

```python
# get the counts of each quality rating 
counts = data['quality'].value_counts()

# create a pie chart
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(counts)
ax.set_title('Distribution of Wine Quality Ratings')
ax.legend(bbox_to_anchor=(1.2, 1), labels=["{}: {:.1f}%".format(label, value/sum(counts)*100) for label, value in zip(counts.index, counts)])

plt.show()
```
